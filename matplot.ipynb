```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
# 按照 X軸座標
x = [1, 2, 3]
y = [4, 4, 6]
# (1,4) (2,4) (3,6)
plt.plot(x, y)
```

```python
# 換順序
x = [1, 3, 2]
y = [4, 6, 4]
# (1,4) (3,6) (2,4) 
plt.plot(x, y)
```

```python
x = [1, 2, 3]
y = [4, 4, 6]
# (1,4) (2,4) (3,6)
plt.scatter(x, y, color='yellow')
```

```python
x = [1, 3, 2]
y = [4, 6, 4]
plt.scatter(x, y, color='red')  # (1,4)(3,6)(2,4)
plt.scatter(y, x, color='blue') # (4,1)(6,3)(4,2)
plt.plot(y, x, color='black') # (4,1)(6,3)(4,2)

# plt.savefig('./scatter_temp.jpg')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend(["X-Y", "Y-X", "Y-X black"])
plt.title("X and Y picture")
plt.savefig('./scatter_red.jpg')
```
